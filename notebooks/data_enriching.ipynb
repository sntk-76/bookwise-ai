```python
import pandas as pd
import numpy as np
import os
import requests
import time
```

```python
data = pd.read_csv('/home/sinatavakoli284/bookwise-ai/data/cleaned_raw_data.csv')
```

```python
data.drop(columns='Unnamed: 0',inplace=True)
data.reset_index(drop=True,inplace=True)
data.info()
```

```python
def description_extractor(data, api_key):
    for index, row in data.iterrows():
        try:

            title = row['title'].replace(" ", "+")
            author = row['authors'].replace(" ", "+")
            query = f"https://www.googleapis.com/books/v1/volumes?q=intitle:{title}+inauthor:{author}&key={api_key}"

            response = requests.get(query)
            response_data = response.json()


            if 'items' in response_data and 'volumeInfo' in response_data['items'][0]:
                description = response_data['items'][0]['volumeInfo'].get('description')
                data.at[index, 'description'] = description


            time.sleep(0.3)

        except Exception as e:
            print(f"Failed at index {index}: {e}")
            data.at[index, 'description'] = None

    return data
```
